```python
import cv2
import dlib
```

```python
from imutils import face_utils
```

```python
cap = cv2.VideoCapture(0)
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
sleepy = 0
drowsy = 0
active = 0
status = ""

color = (255,255,255)

def compute_distance(A, B):
    x, y = A - B
    len = x**2 + y**2
    return len**0.5

def blinked(a,b,c,d,e,f):
    up = compute_distance(b, d) + compute_distance(c, e)
    down = 2*compute_distance(a, f)
    ratio = up/down
    if ratio > 0.25:
        return 2
    elif( ratio>0.2 and ratio <= 0.25):
        return 1
    return 0
```

```python
while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)
    print(type(faces))
    for face in faces:
        x=face.left()
        y=face.top()
        X=face.right()
        Y=face.bottom()
        
        frame_copy = frame.copy()
        cv2.rectangle(frame_copy, (x,y), (X,Y), (255,255,255), 2)
        points = predictor(gray, face)
        points = face_utils.shape_to_np(points) #changes to a numpy array
        left_blink = blinked(points[36], points[37], points[38], points[41], points[40], points[39])
        right_blink= blinked(points[42], points[43], points[44], points[47],points[46],points[45])
        
        if(left_blink == 0 or right_blink == 0):
            sleep+=1
            drowsy=0
            active =0
            if(sleep==12):
                status = "driver needs a break *sleeping"
                color = (255,0,0)
        elif(left_blink==1 or right_blink==1):
            sleep=0
            active=0
            drowsy+=1
            if(drowsy == 6):
                status="drowsy driver"
                color = (0,0,255)
        else :
            drowsy = 0
            sleep = 0
            active+=1
            if(active==3):
                status = "active"
                color=(0,255,0)
        cv2.putText(frame, status, (100,100), cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 2)
        
        for n in range(68):
            (x,y) = points[n]
            cv2.circle(frame_copy, (x,y), 1, (255,0,0), 1)
        cv2.imshow('dots', frame_copy)
    cv2.imshow("processed",frame)
    
    key = cv2.waitKey(1)
    if key == 27:
        break
```
